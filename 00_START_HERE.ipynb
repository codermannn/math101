## The Hook
What is mathematical intuition? It's not memorizing formulas. It's the physical feeling you get when you twist a dial and watch a system react.

Run the cell below. Don't worry about reading the code. **Just grab the sliders.**

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider
%matplotlib inline

@interact(
    frequency=FloatSlider(value=3.0, min=1.0, max=10.0, step=0.5, description='Frequency'),
    phase=FloatSlider(value=0.0, min=0.0, max=2*np.pi, step=0.1, description='Phase'),
    decay=FloatSlider(value=0.5, min=0.0, max=2.0, step=0.1, description='Decay')
)
def mesmerize(frequency, phase, decay):
    fig, ax = plt.subplots(figsize=(10, 5))
    x = np.linspace(0, 10, 500)
    y = np.sin(frequency * x + phase) * np.exp(-decay * x)
    
    ax.plot(x, y, color='#6366f1', linewidth=3)
    ax.fill_between(x, y, alpha=0.2, color='#6366f1')
    ax.axhline(0, color='gray', linewidth=1)
    ax.set_ylim(-1.5, 1.5)
    ax.set_facecolor('#fafafa')
    ax.set_title("Mathematics is a machine. You are the operator.", fontsize=14, fontweight='bold')
    plt.tight_layout()
    plt.show()
```
