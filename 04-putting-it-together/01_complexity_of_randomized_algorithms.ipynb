```python
import numpy as np
import matplotlib.pyplot as plt
from math101.algorithms import quicksort_count, merge_sort_count, insertion_sort_count
%matplotlib inline
```

## The Sorting Algorithm Landscape

Let's compare deterministic vs randomized algorithms across different input types.

```python
sizes = np.arange(50, 501, 50)
input_types = {
    'sorted': lambda n: list(range(n)),
    'reversed': lambda n: list(range(n, 0, -1)),
    'random': lambda n: list(np.random.permutation(n)),
}

fig, axes = plt.subplots(1, 3, figsize=(16, 5))

for ax, (input_name, gen_func) in zip(axes, input_types.items()):
    # MergeSort (deterministic, always O(n log n))
    merge_comps = [merge_sort_count(gen_func(n))[1] for n in sizes]
    
    # Insertion sort (deterministic, O(n) to O(n²))
    insert_comps = [insertion_sort_count(gen_func(n))[1] for n in sizes]
    
    # QuickSort random pivot (randomized, expected O(n log n))
    # Average over 10 runs
    quick_comps = [np.mean([quicksort_count(gen_func(n), random_pivot=True) 
                           for _ in range(10)]) for n in sizes]
    
    # QuickSort first pivot (deterministic)
    quick_det = [quicksort_count(gen_func(n), random_pivot=False) for n in sizes]
    
    ax.plot(sizes, merge_comps, 'o-', color='#10b981', linewidth=2, label='MergeSort', markersize=4)
    ax.plot(sizes, quick_comps, 's-', color='#6366f1', linewidth=2, label='QuickSort (random)', markersize=4)
    ax.plot(sizes, quick_det, '^-', color='#f97316', linewidth=2, label='QuickSort (first)', markersize=4)
    ax.plot(sizes, insert_comps, 'D-', color='#f43f5e', linewidth=2, label='InsertionSort', markersize=4)
    
    ax.set_xlabel('n'); ax.set_ylabel('Comparisons')
    ax.set_title(f'Input: {input_name}', fontsize=12, fontweight='bold')
    ax.legend(fontsize=8); ax.grid(True, alpha=0.3)

plt.suptitle('Algorithm × Input Type: Who wins?', fontsize=14, fontweight='bold', y=1.03)
plt.tight_layout(); plt.show()
```

## Key Insight: Worst-Case vs Expected-Case

| Algorithm | Worst case | Expected case | Best case |
|-----------|-----------|--------------|----------|
| MergeSort | O(n log n) | O(n log n) | O(n log n) |
| QuickSort (first) | O(n²) | O(n log n) | O(n log n) |
| QuickSort (random) | O(n²) | O(n log n) | O(n log n) |
| InsertionSort | O(n²) | O(n²) | O(n) |

Randomized QuickSort has the **same worst case** as deterministic, but no input can **force** it.
The adversary controls the data, but not your coin flips.

```python
# Distribution of runtimes: randomized QuickSort vs deterministic MergeSort
n = 300
trials = 500
arr = list(range(n))  # sorted — worst case for det. QS

qs_comps = [quicksort_count(arr, random_pivot=True) for _ in range(trials)]
ms_comp = merge_sort_count(arr)[1]  # Same every time

fig, ax = plt.subplots(figsize=(9, 5))
ax.hist(qs_comps, bins=40, color='#6366f1', alpha=0.7, edgecolor='white', 
        density=True, label='QuickSort (random) — varies')
ax.axvline(ms_comp, color='#10b981', linewidth=3, linestyle='--', 
           label=f'MergeSort — always {ms_comp}')
ax.axvline(np.mean(qs_comps), color='#f97316', linewidth=2, 
           linestyle=':', label=f'QS mean: {np.mean(qs_comps):.0f}')

ax.set_xlabel('Comparisons'); ax.set_ylabel('Density')
ax.set_title(f'n={n}: Randomized variance vs deterministic certainty', 
            fontsize=13, fontweight='bold')
ax.legend(fontsize=9); ax.grid(True, alpha=0.3)
plt.tight_layout(); plt.show()

print(f'QuickSort (random): mean={np.mean(qs_comps):.0f}, std={np.std(qs_comps):.0f}')
print(f'MergeSort:          always {ms_comp}')
print(f'QuickSort is faster ON AVERAGE but has variance.')
```
