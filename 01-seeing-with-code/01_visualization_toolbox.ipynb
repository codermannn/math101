## 1. Concept Intuition

Think of matplotlib like a **physical drawing desk**:

- The **Figure** is the paper you're drawing on
- The **Axes** is a framed region where one specific chart lives
- **Lines, patches, text** are things you draw inside that frame
- The **Axis** (x-axis, y-axis) is the ruler on the edge

Key insight: **everything is an object**. When you call `ax.plot(...)`, you're not creating a picture — you're creating a `Line2D` Python object that *lives inside* the Axes object. You can grab it, inspect it, change its color, swap its data.

NumPy is your **calculator**. It doesn't care about plotting — it just creates and transforms arrays of numbers blazingly fast. Matplotlib then *reads* those arrays and draws them.

```
NumPy: creates the data    →  [0, 0.1, 0.2, ..., 6.28]
                                         ↓
Matplotlib: draws the data →  a smooth sine curve on screen
```

## 2. Visual Representation

Let's see the object hierarchy in action.

````python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Create the canvas (Figure) and one drawing area (Axes)
fig, ax = plt.subplots(figsize=(8, 4))

# Generate data with NumPy
x = np.linspace(0, 2 * np.pi, 200)
y_sin = np.sin(x)
y_cos = np.cos(x)

# Plot returns Line2D objects — capture them!
line_sin, = ax.plot(x, y_sin, label='sin(x)', color='#6366f1', linewidth=2)
line_cos, = ax.plot(x, y_cos, label='cos(x)', color='#f97316', linewidth=2, linestyle='--')

ax.set_title('Two waves on one Axes', fontsize=14, fontweight='bold')
ax.legend()
ax.grid(True, alpha=0.3)
ax.set_xlabel('x (radians)')
ax.set_ylabel('amplitude')

plt.tight_layout()
plt.show()
````

## 3. Mathematical Formulation

The data pipeline is simple algebra:

Given a domain $x \in [a, b]$ sampled at $n$ points:

$$x = \text{linspace}(a, b, n) = \left[a,\; a + \frac{b-a}{n-1},\; a + 2\cdot\frac{b-a}{n-1},\; \ldots,\; b\right]$$

Then $y = f(x)$ is computed **element-wise** by NumPy — no loops needed.

This is **vectorization**: applying a function to an entire array at once. It's the reason NumPy is fast and the foundation of everything we'll visualize.

## 4. Code Implementation

### The Figure/Axes object model

````python
# Multiple Axes on one Figure — the subplots grid
fig, axes = plt.subplots(1, 3, figsize=(12, 3))

x = np.linspace(0, 4, 100)
functions = [('Linear: y=x', lambda x: x, '#10b981'),
             ('Quadratic: y=x²', lambda x: x**2, '#6366f1'),
             ('Cubic: y=x³', lambda x: x**3, '#f43f5e')]

for ax, (title, f, color) in zip(axes, functions):
    ax.plot(x, f(x), color=color, linewidth=2)
    ax.set_title(title, fontsize=11)
    ax.grid(True, alpha=0.3)

fig.suptitle('One Figure, Three Axes', fontsize=14, fontweight='bold', y=1.05)
plt.tight_layout()
plt.show()
````

````python
# Modifying plots AFTER creation
fig, ax = plt.subplots(figsize=(6, 4))
x = np.linspace(0, 5, 100)
line, = ax.plot(x, np.exp(x), color='gray', linewidth=1)

# Now modify the line object directly
line.set_color('#8b5cf6')
line.set_linewidth(3)
line.set_linestyle('-.')
ax.set_title('Modified after creation', fontsize=13)
ax.set_facecolor('#fafafa')
plt.show()

# Proof: inspect the object
print(f'Type: {type(line)}')
print(f'Color: {line.get_color()}')
print(f'Linewidth: {line.get_linewidth()}')
print(f'Data shape: x={line.get_xdata().shape}, y={line.get_ydata().shape}')
````

### Simple Animation

Animation = redrawing a plot with updated data in a loop.

````python
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

fig, ax = plt.subplots(figsize=(7, 4))
x = np.linspace(0, 2 * np.pi, 200)
line, = ax.plot(x, np.sin(x), color='#6366f1', linewidth=2)
ax.set_ylim(-2.5, 2.5)
ax.set_title('Growing sine wave', fontsize=13)
ax.grid(True, alpha=0.3)

def update(frame):
    amplitude = 0.1 + frame * 0.05
    line.set_ydata(amplitude * np.sin(x))
    ax.set_title(f'Amplitude = {amplitude:.2f}', fontsize=13)
    return line,

ani = FuncAnimation(fig, update, frames=40, interval=80, blit=True)
plt.close(fig)
HTML(ani.to_jshtml())
````

## 5. Interactive Experiment

Use the sliders to control the wave. Watch how **frequency** and **amplitude** change the shape independently.

````python
from ipywidgets import interact, FloatSlider

@interact(
    amplitude=FloatSlider(value=1.0, min=0.1, max=3.0, step=0.1, description='Amplitude'),
    frequency=FloatSlider(value=1.0, min=0.5, max=5.0, step=0.5, description='Frequency'),
    phase=FloatSlider(value=0.0, min=0, max=2*np.pi, step=0.1, description='Phase')
)
def wave_lab(amplitude, frequency, phase):
    fig, ax = plt.subplots(figsize=(8, 4))
    x = np.linspace(0, 4 * np.pi, 300)
    y = amplitude * np.sin(frequency * x + phase)
    ax.plot(x, y, color='#6366f1', linewidth=2)
    ax.fill_between(x, y, alpha=0.1, color='#6366f1')
    ax.set_ylim(-3.5, 3.5)
    ax.set_title(f'y = {amplitude:.1f} · sin({frequency:.1f}x + {phase:.1f})', fontsize=13)
    ax.axhline(y=0, color='gray', linewidth=0.5)
    ax.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()
````

## 6. Tool Exploration

Run each cell below to **inspect the objects** you've been using. This is how you learn any library — by looking inside.

````python
# Explore the Figure object
fig, ax = plt.subplots()
ax.plot([1, 2, 3], [1, 4, 9])
plt.close(fig)

print('=== Figure ===')
print(f'Type: {type(fig)}')
print(f'Size: {fig.get_size_inches()}')
print(f'Axes inside: {fig.get_axes()}')
print()

print('=== Axes ===')
print(f'Type: {type(ax)}')
print(f'Lines: {ax.lines}')
print(f'X limits: {ax.get_xlim()}')
print(f'Y limits: {ax.get_ylim()}')
print()

print('=== Line2D ===')
line = ax.lines[0]
print(f'Type: {type(line)}')
print(f'Data: x={line.get_xdata()}, y={line.get_ydata()}')
print(f'Color: {line.get_color()}')
print()

# Try this yourself:
# print([m for m in dir(line) if m.startswith('get_')])
````

````python
# Explore NumPy arrays
arr = np.linspace(0, 10, 5)
print(f'Array: {arr}')
print(f'Type: {type(arr)}')
print(f'Shape: {arr.shape}')
print(f'Dtype: {arr.dtype}')
print(f'Vectorized sin: {np.sin(arr)}')
print()
print('Key insight: np.sin(array) applies to ALL elements at once — no loop needed.')
````
